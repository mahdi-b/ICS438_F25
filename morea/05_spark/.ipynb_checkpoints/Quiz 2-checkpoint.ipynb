1. Any problem can be solved using a MapReduce approach.
  * A - True
  * B - False: 




B - False
Not every problem can be solved (or at least efficiently solved) using a MapReduce approach. 
While MapReduce is powerful for certain types of distributed data processing tasks, it has limitations and isn't appropriate
for all problems, especially those requiring complex algorithms or frequent data dependencies.
Reacall the  jigsaw puzzle example? Each student is given a subset of the puzzle pieces. This setup is analogous to a distributed computing environment, where different nodes (our students) each have a portion of the data (puzzle pieces). Here are some of the issues with this approach:
1. No single student has a complete view of the puzzle, making it difficult to understand the overall picture.
2. A lot of time might be spent describing pieces to each other so that can exchange them rather than assembling them.
3. Some students might have many easily connectable pieces, while others have few or difficult pieces.
  * The process might require many rounds of sharing and collaboration, which isn't ideal for standard MapReduce.

* When will Spark actually execute the transformations (map and filter) in the lazy execution example?


  * A) Immediately after calling the map function.
  * B) Immediately after calling the filter function.
  * C) Only when the collect() action is called.
  * D) As soon as the parallelize() function is called.


C) Only when the collect() action is called.
The answr is in the quesiton. Spark uses lazy evaluation. Transformations (like map and filter) are not executed immediately. They are only executed when an action (like collect()) is called.

7. Is the line `rdd_filtered = rdd.filter(lambda x: x > 25)` an action or a transformation in Spark?
  * A) Action
  * B) Transformation


B) Transformation
The filter operation is a transformation in Spark. It build "an execution plan to" transform the data by filtering values that are <= 25 and creates a new RDD without modifying the original one. Not that this function doesn't trigger any computation on its own.

10. How many stages will Spark create to execute the following code?
```python
rdd = sc.parallelize([("a", 1), ("b", 2), ("a", 3), ("b", 4)])
rdd_mapped = rdd.map(lambda x: (x[0], x[1] * 2))
rdd_reduced = rdd_mapped.reduceByKey(lambda a, b: a + b)
result = rdd_reduced.collect()
```
  * A) 1 stage
  * B) 2 stages
  * C) 3 stages
  * D) 4 stages

B) 2 stages
Recall that each stage represents a set of transformations that can be executed in parallel without requiring shuffling between partitions.
Therefore, this code will create 2 stages:







1. **Stage 1 (parallelize + map)**:
- `parallelize`: Converts the input list into an RDD. This is a parallel operation that doesn't involve shuffling.
- `map`: The `map` transformation applies a function to each element independently and doesn't require any data movement across partitions. Therefore, it happens within the same stage.

2. **Stage 2 (reduceByKey)**:
- reduceByKey requires a shuffle because it needs to group the keys ("a" and "b") together from different partitions. This operation moves  data across the cluster, which triggers a new stage.
- The action collect() is what triggers the actual execution, but it doesn’t create a separate stage by itself. It simply triggers the computation.
